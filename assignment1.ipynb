```python
# fetching the positive,Negative and Neutal txt files and storing it in a list
pos=[]

with open("D:/Users/kpgaj/Documents/positive.txt",errors='ignore') as f:
        for i in f: 
            pos.append(i.rstrip("\n")) 
             
neg=[]
with open("D:/Users/kpgaj/Documents/negative.txt",errors='ignore') as f:
        for i in f: 
            neg.append(i.rstrip("\n")) 
            
neutral=[]
with open("D:/Users/kpgaj/Documents/neutral.txt",errors='ignore') as f:
        for i in f: 
            neutral.append(i.rstrip("\n")) 
```

```python
#lower case conversion 

Sample['text']=Sample['text'].str.lower()
```

```python
#removing the punctuatioins 
import string
table = str.maketrans('', '', string.punctuation)
Sample['text_list']= [w.translate(table) for w in Sample['text']]
```

```python
#splitting the cleaned text
Sample['text_list']=Sample['text_list'].str.split()
```

```python
Sample['text_list']
```

```python
#vectoring the cleaned text data 
from sklearn.feature_extraction.text import TfidfVectorizer

vectorizer = TfidfVectorizer()
X = vectorizer.fit_transform(Sample['text'])
dictionary=vectorizer.get_feature_names()

  

```

```python
#comparing if the get_feature_names elements are present in the three list and storing the elements in a new list of POS NEG and NEUTRAl
POS=[]
NEG=[]
NEUTRAL=[]
for i in dictionary:
    if i in pos :
        POS.append(i)
    if i in neg:
        NEG.append(i)
    if i in neutral:
        NEUTRAL.append(i)
```

```python
#creating a new column called status which will have pos,neg and neutral labels
Sample['status']='None'
```

```python
Sample['text']
```

```python
#comparing the elements of the POS Neg and NEUTRAL with those of the splitted text list .
pcount =0
ncount=0
neuc=0

for rowt,rowl in zip(Sample['text_list'],range(len(Sample['status']))):
    #print(rowl)
    for i in rowt:
        i =i.rstrip("\n")
        i = i.lower()
        if i in POS:
            pcount = pcount+1
        if i in NEG:
            ncount = ncount+1
        if i in NEUTRAL:
            neuc = neuc+1
    if pcount > ncount and pcount > neuc :
        Sample['status'][rowl] = "pos"
        
    if ncount > pcount and ncount > neuc:
        Sample['status'][rowl] = "neg"
       
    if neuc > ncount and neuc > pcount:
        Sample['status'][rowl] = "neutral"
      
    if pcount == ncount:
        Sample['status'][rowl] = "neutral"
        
    print(rowl)# just to keep a count if the labelling is hitting the column
    
    pcount =0
    ncount=0
    neuc=0
```

```python
#preparing data for fitting in the sklearn models
count_vect = CountVectorizer()
X_train_counts = count_vect.fit_transform(X_train)
tfidf_transformer = TfidfTransformer()
X_train_tfidf = tfidf_transformer.fit_transform(X_train_counts.astype(str))
```

```python
#chossing ada booster classifier
from sklearn.metrics import confusion_matrix
from sklearn.ensemble import AdaBoostClassifier
from sklearn.model_selection import cross_val_score
from sklearn.metrics import accuracy_score
from sklearn.metrics import classification_report
clf =  AdaBoostClassifier()
clf=clf.fit(X_train_tfidf, y_train)

predict = clf.predict(count_vect.transform(X_test))

print("Confusion Matrix for Ada Booster Classifier:")
print(confusion_matrix(y_test,predict))
print("Score:",round(accuracy_score(y_test,predict)*100,2))
print("Classification Report:" + "\n",classification_report(y_test,predict))
print("cross validation Score",cross_val_score(clf, X_train_tfidf, y_train, cv=3)) 

```
